```python
import NormalizationModel as nm 
import numpy as np
import matplotlib.pyplot as plt

n_it = 180
tar_x = np.round(np.random.normal(0, 5, n_it))
tar_y = np.round(np.random.normal(10, 5, n_it))
tar_clr = np.round(np.random.uniform(-80,-60, n_it))

dist_x = np.round(np.random.normal(0, 5, n_it))
dist_y =  np.round(np.random.normal(10, 5, n_it))
dist_clr = np.round(np.random.uniform(80,60, n_it))
dist_clr =  np.ones(n_it)*70

# att_x = [None, None, -70]

mdl = nm.Model(
    stimulus_center = np.vstack(
        [np.hstack([tar_x,dist_x]),
         np.hstack([tar_y,dist_y]),
         np.hstack([tar_clr,dist_clr])]).T,
    impulse_width = [1,1,1],
    EF_width=[3,30],
    attention_center = [None, None, -70],
    AF_width = [5, 120],
    position_span=50
)
```

```python
mdl.stimulus.plot();
```

```python
mdl.attention.plot();
```

```python
mdl.stimulus_drive.plot();
```

```python
mdl.suppressive_drive.plot();
```

```python
mdl.response.plot();
```
